```python
# Calculating RFM metrics 
# Before proceeding with the calculations, let's create the date that represents the day we will analyze
df["last_order_date"].max()  # to find out when we last shopped (2021-05-30)
today_date = dt.datetime(2021,6,1) # We created the day for two days after the last shopping
```

```python
# calculating recency,frequency,monetary values 
rfm = pd.DataFrame() #Let's create a dataframe to insert the values we will create into it

rfm["customer_id"] = df["master_id"] # adding customer numbers 
rfm["recency"] = (today_date - df["last_order_date"]).dt.days # Calculate how many days have passed since your last purchase
rfm["frequency"] = df["order_num_total"] # expressing the frequency of shopping 
rfm["monetary"] = df["customer_value_total"] # total expenditure  
rfm.head()
```

```python
# Calculation of RFM scores 
# We score it by dividing it into 5 equal parts with the qcut function. A small recency value indicates that the customer has shopped recently, so we score backwards from 5.   
rfm["recency_score"] = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])

rfm["frequency_score"] = pd.qcut(rfm['frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm["monetary_score"] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])
rfm.head()
```

```python
# Calculating the RF score 
rfm["RF_SCORE"] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
```

```python
# To make Rf scores more explainable, let's segment customers according to their scores
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'}  # according to the seg_map dictionary we can do this

# The regex argument is used to replace values in ranges, so that 5[0-5] is changed to champions, which matches values between 50 and 55.
rfm['segment'] = rfm['RF_SCORE'].replace(seg_map, regex=True)
rfm.head()
```

```python
# If we want to examine the segments in more detail  
rfm[["segment", "recency", "frequency", "monetary"]].groupby("segment").agg(["mean", "count"])
```

```python
# Let's find customers in the relevant profile with the help of RFm analysis.
# FLO is adding a new women's footwear brand. The brand's product prices are above the general customer preferences. Therefore, the brand's
# We want to contact customers with a profile that would be interested in product promotion and sales. We want these customers to be loyal and
# It is planned to be customers who shop in the women's category. Put customers' ID numbers in csv file yeni_marka_hedef_mteri_id.csv
# Save as.
# Let's select loyal customers from segments 
target_segments_customer_ids = rfm[rfm["segment"].isin(["champions","loyal_customers"])]["customer_id"] #segmnetler ierisinden sadk mterileri seelim 

# ids of loyal and female category shoppers
cust_ids = df[(df["master_id"].isin(target_segments_customer_ids)) &(df["interested_in_categories_12"].str.contains("KADIN"))]["master_id"] 

cust_ids.to_csv("yeni_marka_hedef_mteri_id.csv", index=False) # save data with customer ids as csv file
cust_ids.shape # If we want to take a quick look at how many people there are 
```
